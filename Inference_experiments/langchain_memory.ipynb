# Imports

```python
from langchain.llms import HuggingFacePipeline
```

```python
from transformers import AutoTokenizer
```

```python
import transformers
import torch
```

# Load Llama2 model

```python
model="meta-llama/Llama-2-7b-chat-hf"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
pipeline=transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
    )
```

```python
llm=HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature':0})
```

```python
prompt="What is finetuning?"
```

```python
print(llm(prompt))
```

```python
print(llm("tell me a finetuning joke"))
```

```python
print(llm("how about another one"))
```

```python
print(llm("you tell me?"))
```

# Imports for PromptTemplates

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
```

```python
prompt_template_name = PromptTemplate(
    input_variables =['product'],
    template = "What is a good name for a company that makes {product}"
)
```

```python
chain = LLMChain(llm=llm,prompt=prompt_template_name)
name = chain.run("Finetuned LLM models")
print(name)
```

```python
prompt_template_name = PromptTemplate(
    input_variables =['product'],
    template = "What would be  a good name for a company that makes {product}"
)
```

```python
chain = LLMChain(llm=llm,prompt=prompt_template_name)
name = chain.run("Large language models security and safety")
print(name)
```

```python
from langchain.memory import ConversationBufferMemory

memory = ConversationBufferMemory()

prompt_template_name = PromptTemplate(
    input_variables =['product'],
    template = "What is a good name for a company that makes {product}"
)

chain = LLMChain(llm=llm, prompt=prompt_template_name, memory=memory)
name = chain.run("Vector embeddings")
print(name)
```
