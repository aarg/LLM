# Imports

````python
from langchain.llms import HuggingFacePipeline
````

````python
from transformers import AutoTokenizer
````

````python
import transformers
import torch
````

# Load llama2 model

````python
model="meta-llama/Llama-2-7b-chat-hf"
````

````python
tokenizer=AutoTokenizer.from_pretrained(model)
````

````python
pipeline=transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
    )
````

````python
llm=HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature':0})
````

````python
prompt="What is finetuning?"
````

````python
print(llm(prompt))
````

````python
customer_email = """
Arrr, I be fuming that me blender lid \
flew off and splattered me kitchen walls \
with smoothie! And to make matters worse,\
the warranty don't cover the cost of \
cleaning up me kitchen. I need yer help \
right now, matey!
"""
````

````python
style = """American English \
in a calm and respectful tone
"""
````

````python
prompt = f"""Translate the text \
that is delimited by triple backticks
into a style that is {style}.
text: ```{customer_email}```
"""

print(prompt)
````

````python
print(llm(prompt))
````

# Let's use langchain prompts

````python
from langchain.prompts import PromptTemplate
````

````python
from langchain.chains import LLMChain
````

### Prompt template

````python
template_string = """Translate the text \
that is delimited by triple backticks \
into a style that is {style}. \
text: ```{text}```
"""
````

````python
customer_style = """American English \
in a calm and respectful tone
"""
````

````python
customer_email = """
Arrr, I be fuming that me blender lid \
flew off and splattered me kitchen walls \
with smoothie! And to make matters worse, \
the warranty don't cover the cost of \
cleaning up me kitchen. I need yer help \
right now, matey!
"""
````

````python
prompt_template=PromptTemplate(input_variables=["style", "text"],
                               template=template_string)
````

````python
chain = LLMChain(llm=llm, prompt=prompt_template, verbose=True)
````

````python
response = chain.run({'style': customer_style, 'text': customer_email})
````

````python
print(response)
````

## Translate to English

````python
template_string = """Translate the text \
that is delimited by triple backticks \
into English language starting with "English text:" \
text: ```{text}```
"""
````

````python
customer_string = """Sous le pont Mirabeau coule la Seine
Et nos amours
Faut-il qu’il m’en souvienne
La joie venait toujours après la peine
"""
````

````python
prompt_template=PromptTemplate(input_variables=["text"],
                               template=template_string)
chain = LLMChain(llm=llm, prompt=prompt_template, verbose=True)
response = chain.run({'text': customer_string})
print(response)
````
